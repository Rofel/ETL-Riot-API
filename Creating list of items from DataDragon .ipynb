```python
import pandas as pd
import json
```

```python
json_file_path = 'item.json'  # Replace with the path to your JSON file
with open(json_file_path, 'r', encoding='utf-8') as json_file:
    item_data = json.load(json_file)
```

```python
item_data.keys()
```

```python
type(item_data['data'])
```

```python
key_list = list(item_data['data'].keys())
```

```python
key_list
```

```python
item_list = []
for item in key_list:
    item_ = []
    item_id = item
    item_name = item_data['data'][item]['name']
    item_cost = item_data['data'][item]['gold']['base']
    item_.extend([item_id, item_name,item_cost])
    item_list.append(item_)
```

```python
item = key_list[0]
```

```python
item
```

```python
item_data['data'][item]
```

```python
item_list
```

```python
items_df = pd.DataFrame(item_list, columns=['itemId','itemName','itemCost'])
```

```python
items_df
```

```python
items_df.to_csv('items.csv', index=False)
```
