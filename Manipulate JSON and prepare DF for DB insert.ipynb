#### Populating matchs_df with the data from matches

```python

for match in match_data:
    #matchs_df info
    
    match_info = matchs_df_populate(match)
    
    match_ = pd.Series(match_info, index=matchs_df.columns)
    matchs_df = pd.concat([matchs_df, match_.to_frame().T], ignore_index=True)
```

## Player Stats Data

```python
for match in match_data:
    #matchs_df info
    for i,participant in enumerate(match['info']['participants']):
        player_stats = player_stats_populate(match,participant)

        match_ = pd.Series(player_stats, index=players_stats_df.columns)
        players_stats_df = pd.concat([players_stats_df, match_.to_frame().T], ignore_index=True)
```

## Bans Data

```python




for match in match_data:
    for i, team in enumerate(match['info']['teams']):
        bans_info =[]
        matchid = match['metadata']['matchId']
        teamId = team['teamId']
        bans_info.extend([matchid,teamId])
        for i, ban in enumerate(team['bans']):
            ban= ban['championId']
            bans_info.extend([ban])

        bans_ = pd.Series(bans_info, index=bans_df.columns)
        bans_df = pd.concat([bans_df, bans_.to_frame().T], ignore_index=True)    
```

```python
bans_df
```

```python
bans_df = bans_df.drop_duplicates()
```

## Teams Data

```python
for match in match_data:
    for i, team in enumerate(match['info']['teams']):
            team_info= team_data_populate(match, team)

            team_ = pd.Series(team_info, index=teams_df.columns)
            teams_df = pd.concat([teams_df, team_.to_frame().T], ignore_index=True)   


```

```python
teams_df =teams_df.drop_duplicates()
```

```python
teams_df
```

## Summoner Data

```python
for match in match_data:
    for participant in match['info']['participants']:
        summoner_info =[]
        
        puuid = participant['puuid']
        summonerId = participant['summonerId']
        summonerName = participant['summonerName']
        riotIdGameName =participant['riotIdGameName']
        riotIdTagline = participant['riotIdTagline']
        
        summoner_info.extend([puuid,summonerId,summonerName,riotIdGameName,riotIdTagline])
        
        player_ = pd.Series(summoner_info, index=players_df.columns)
        players_df = pd.concat([players_df, player_.to_frame().T], ignore_index=True)   
        
        
```

```python
players_df = players_df.drop_duplicates('puuid')
```
